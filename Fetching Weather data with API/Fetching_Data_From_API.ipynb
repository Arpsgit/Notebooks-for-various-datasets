## Fetching Data using Rapid API site which provides several APIs

 I fetch this data from Visual Crossing Weather's API with the help of Rapid API and then I processed it according to my requirement and automate the task of fetching the data between two certain dates using python and then I export that into a CSV.

```python
import datetime
start = datetime.date(2017,1,1)
end = datetime.date(2018,12,31)
df = pd.DataFrame()
while start < end:
    url = "https://visual-crossing-weather.p.rapidapi.com/history"
    headers = {
            "X-RapidAPI-Key": "e0089e0c2emsh0bebbd1bc5341b2p1042d6jsn3e48e26fe9c4",
            "X-RapidAPI-Host": "visual-crossing-weather.p.rapidapi.com"
        } # provide your own X-RapidAPI-Key and X-RapidAPI-Host from rapid api site - Visual Crossing Weather
    if end > start + datetime.timedelta(29):
        
        start_date = start
        start_str = str(start_date) + 'T00:00:00'
        End_date = start_date + datetime.timedelta(29) # as this api gives only 30 days' data in a single request, so I made the end date according to that rule.
        End_str =  str(End_date) + 'T00:00:00'
        querystring = {f"startDateTime":{start_str},"aggregateHours":"24","location":"Kolkata,West Bengal,India","endDateTime":{End_str},"unitGroup":"uk","contentType":"csv","shortColumnNames":"0"}
    else:
        start_date = start
        start_str = str(start_date) + 'T00:00:00'
        End_date = end
        End_str =  str(End_date) + 'T00:00:00'
        querystring = {f"startDateTime":{start_str},"aggregateHours":"24","location":"Kolkata,West Bengal,India","endDateTime":{End_str},"unitGroup":"uk","contentType":"csv","shortColumnNames":"0"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    temp_df = pd.read_csv(StringIO(response.text)) # made that reponse text into a comma separated value and then it was turned into a data frame.
    df = df.append(temp_df, ignore_index= True) # merge those every response's data to make a single data frame.
    start = End_date + datetime.timedelta(1)
```
