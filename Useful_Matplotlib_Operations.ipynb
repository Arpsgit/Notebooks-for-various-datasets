# Bar Plot
<br> For bivariate analysis

```python
#data
#num-cat cols
year=[2016,2017,2018,2019,2020,2021,2022,2023]
marks=[9.7,7,8.8,6.5,7.2,8.1,9.1,9.5]
```

```python
# scatter plot of tips w.r.t total bill where the color denotes the sex of the customer
# This time with legend
plt.grid(visible = True)
plt.scatter(df_tips[df_tips['sex']==1]['total_bill'], df_tips[df_tips['sex']==1]['tip'], c='red', alpha=0.75, label='Male')
plt.scatter(df_tips[df_tips['sex']==0]['total_bill'], df_tips[df_tips['sex']==0]['tip'], c='green', alpha=0.75, label='Female')
plt.xlabel('Total amount of bill')
plt.ylabel('Tips given')
plt.title('Analysing Tips with respect to Total bill')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
z1 = np.sin(xx) + np.cos(yy)
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(projection = '3d')

ax.plot_surface(xx,yy,z, cmap='viridis')
fig.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(projection = '3d')

ax.plot_surface(xx,yy,z1, cmap='coolwarm_r')
fig.show()
```

# Contour Plot
<br> It presents a 3D plot on a 2D plane

```python
fig = plt.figure(figsize=(12,10))

ax1 = fig.add_subplot(2,2,1)
ax1.contour(xx,yy,z, cmap='viridis')

ax2 = fig.add_subplot(2,2,2)
ax2.contourf(xx,yy,z, cmap='viridis')

ax3 = fig.add_subplot(2,2,3)
ax3.contour(xx,yy,z1, cmap='viridis')

ax4 = fig.add_subplot(2,2,4)
ax4.contourf(xx,yy,z1, cmap='viridis')

fig.show()
```
